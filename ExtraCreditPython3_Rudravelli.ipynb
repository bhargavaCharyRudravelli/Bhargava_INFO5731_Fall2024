```python
# Import essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
# Load the Bread Basket dataset
sales_data = pd.read_csv("/content/sample_data/bread_basket (1).csv")

print(f"Dataset Shape: {sales_data.shape}\n")
print("Data Columns and Types:")
print(sales_data.dtypes)
```

```python
# Determine the most frequent time segment
dominant_period = sales_data["period_day"].mode()[0]
dominant_count = sales_data["period_day"].value_counts().iloc[0]
print(f"Most active period: {dominant_period} ({dominant_count} transactions)")

# Frequency and ratio by period
period_counts = sales_data["period_day"].value_counts()
period_ratios = sales_data["period_day"].value_counts(normalize=True)
period_summary = pd.DataFrame({"Count": period_counts, "Share": period_ratios})
print("\nTransactions by Time of Day:\n", period_summary)
```

```python
plt.figure(figsize=(6,6))
plt.pie(period_counts, labels=period_counts.index, autopct="%.1f%%", startangle=90)
plt.title("Transactions by Period of Day")
plt.show()
```

```python
# Identify most common items
top_items = sales_data["Item"].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=top_items.index[:15], y=top_items.values[:15], palette="crest")
plt.xticks(rotation=45)
plt.title("Top 15 Sold Items", fontsize=14)
plt.xlabel("Item")
plt.ylabel("Transaction Frequency")
plt.show()
```

```python
selected_items = ["Coffee", "Bread", "Sandwich", "Juice", "Chicken Stew"]
print("Transaction count for selected items:")
for item in selected_items:
    print(f"{item}: {sales_data[sales_data['Item'] == item].shape[0]}")

weekday_ratio = sales_data[sales_data["weekday_weekend"] == "Weekday"].shape[0] / len(sales_data)
print(f"\nWeekday Transaction Share: {weekday_ratio:.2f}")
```

```python
basket_matrix = pd.read_excel("/content/sample_data/encoded_transaction (1).xls")

# Convert to binary (0/1)
basket_matrix = basket_matrix.applymap(lambda x: 1 if x > 0 else 0)
print("Unique encoded values:", np.unique(basket_matrix.values))
```

```python
freq_items_high = fpgrowth(basket_matrix, min_support=0.95, use_colnames=True)
freq_items_low = fpgrowth(basket_matrix, min_support=0.20, use_colnames=True)

print(f"Frequent itemsets (support  0.95): {freq_items_high.shape[0]}")
print(f"Frequent itemsets (support  0.20): {freq_items_low.shape[0]}")
```

```python
assoc_rules = association_rules(freq_items_low, metric="confidence", min_threshold=0.7)

filtered_rules = assoc_rules[assoc_rules["antecedents"].apply(
    lambda s: any(x in s for x in ["Bread","Coffee","Cake","Tea"])
)]

print("Filtered Association Rules:")
display(filtered_rules[["antecedents","consequents","support","confidence","lift","conviction"]])
```

```python
target = filtered_rules[
    (filtered_rules["antecedents"] == {"Coffee", "Cake"}) &
    (filtered_rules["consequents"] == {"Bread"})
]

if not target.empty:
    s, c, l = target["support"].iloc[0], target["confidence"].iloc[0], target["lift"].iloc[0]
    print(f"Rule {{Coffee, Cake}}  {{Bread}}")
    print(f"Support: {s:.3f}, Confidence: {c:.3f}, Lift: {l:.3f}")
    print("Interpretation: Customers buying Coffee and Cake often purchase Bread too.")
else:
    print("Rule {Coffee, Cake}  {Bread} not found.")
```

```python
from sklearn.linear_model import LinearRegression

claims_data = pd.read_excel("/content/sample_data/Claimsdata (2).xlsx")
print("Columns available:", claims_data.columns.tolist())

# Clean Claims column
claims_data["Claims"] = pd.to_numeric(claims_data["Claims"], errors="coerce")
claims_data.dropna(subset=["Claims"], inplace=True)
```

```python
plt.scatter(claims_data["Period"], claims_data["Claims"], color="navy")
plt.title("Claims vs Period")
plt.xlabel("Period")
plt.ylabel("Claims")
plt.show()
```

```python
X = claims_data[["Period"]]
y = claims_data["Claims"]
lr = LinearRegression().fit(X, y)

plt.scatter(X, y, color="steelblue")
plt.plot(X, lr.predict(X), color="darkred", label="Trend Line")
plt.title("Claims Trend and Forecast")
plt.xlabel("Period")
plt.ylabel("Claims")
plt.legend()
plt.show()

print(f"Equation: Claims = {lr.intercept_:.2f} + {lr.coef_[0]:.2f}  Period")
```

```python
future_periods = pd.DataFrame({"Period": range(25, 31)})
future_periods["Predicted_Claims"] = lr.predict(future_periods)
print("\nForecasted Claims (Periods 2530):")
print(future_periods)
```

```python
claims_data["SMA_2"] = claims_data["Claims"].rolling(window=2).mean()

plt.plot(claims_data["Period"], claims_data["Claims"], label="Actual", color="teal")
plt.plot(claims_data["Period"], claims_data["SMA_2"], label="2-Period Average", color="orange")
plt.title("Claims with 2-Period Moving Average")
plt.xlabel("Period")
plt.ylabel("Claims")
plt.legend()
plt.show()
```
